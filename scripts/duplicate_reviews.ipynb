```python
import pandas as pd
```

```python
DATA_PATH = "./drive/MyDrive/drugproject/3_data"

test = DATA_PATH + "/drugsComTest_raw.tsv"
train = DATA_PATH + "/drugsComTrain_raw.tsv"
new = DATA_PATH + "/drugComScrapedData.tsv"
 
new_df = pd.read_csv(new, sep="\t")
test_df = pd.read_csv(test, sep="\t", index_col=0)
train_df = pd.read_csv(train, sep="\t", index_col=0)
```

Grasser et al did not properly clean their dataset. Repeated reviews were not discarded, as the row counts below show.

```python
print("w/ duplicates ||  w/o duplicates")
print(len(new_df), "   ||   ", len(new_df.drop_duplicates('review')))
print(len(test_df), "   ||   ",len(test_df.drop_duplicates('review')))
print(len(train_df), "  ||   ", len(train_df.drop_duplicates('review')))
print()
print("Num rows Grasser et al used: ", sum([len(test_df), len(train_df)]))
print("Num of non-duplicate rows: ", sum([len(test_df.drop_duplicates('review')), len(train_df.drop_duplicates('review'))]))
print("Num non-duplicate rows gotten w scraping: ", len(new_df.drop_duplicates('review')))
print("Num rows in our dataset: ", sum([len(test_df.drop_duplicates('review')), len(train_df.drop_duplicates('review')), len(new_df.drop_duplicates('review'))]))
```

```python
print("Original training set length: {}".format(len(train_df)))
print("Original test set length: {}".format(len(test_df)))
test_dups = len(test_df) - len(test_df.drop_duplicates('review'))

print("{} rows present in testing that are duplicates of training and {} rows in test set are duplicates".format(sum(test_df['review'].isin(train_df['review'])), test_dups))

print("{} rows in training data are duplicates".format(len(train_df) - len(train_df.drop_duplicates("review"))))
```

```python
print(sum(test_df['review'].isin(train_df['review'])))
sum(test_df['review'].isin(train_df.drop_duplicates('review')['review']))
```
