```python
TRAIN_FILE = './drive/MyDrive/drugproject/3_data/train_all_data.csv'
TEST_FILE = './drive/MyDrive/drugproject/3_data/test_all_data.csv'
VALID_FILE = './drive/MyDrive/drugproject/3_data/valid_all_data.csv'
```

```python
tokenizer = get_tokenizer('spacy')
counter_reviews = Counter()

REVIEW_COL_NUM = 2
print("Training tokenizer:\n")
with open(TRAIN_FILE, 'r', encoding="utf8") as f:
    csv_reader = reader(f)
    for i, row in enumerate(csv_reader):
        tokens = tokenizer(row[REVIEW_COL_NUM])
        counter_reviews.update(tokens)
        if not i % 10000:
            print(f"{i} examples completed")

print("Validation tokenizer:\n")
with open(VALID_FILE, 'r', encoding="utf8") as f:
    csv_reader = reader(f)
    for i, row in enumerate(csv_reader):
        tokens = tokenizer(row[REVIEW_COL_NUM])
        counter_reviews.update(tokens)
        if not i % 10000:
            print(f"{i} examples completed")

vocab_reviews = Vocab(counter_reviews, min_freq=25)
```

Writing the vocab to a pickle file

```python
import pickle
path = './drive/MyDrive/drugproject/3_data/vocabs/drug_review_vocab_stoi.pkl'
a_file = open(path, "wb")
pickle.dump(vocab_reviews.stoi, a_file)
a_file.close()
```
